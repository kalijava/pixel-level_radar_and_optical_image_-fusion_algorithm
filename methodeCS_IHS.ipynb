```python
import cv2
import numpy as np

def rgb_to_ihs(image):
    # Convertir l'image RGB en espace de couleur IHS
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    ihs_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    ihs_image[:, :, 2] = cv2.normalize(ihs_image[:, :, 2], None, 0, 255, cv2.NORM_MINMAX)
    return ihs_image

def ihs_to_rgb(ihs_image):
    # Convertir l'image IHS en RGB
    ihs_image[:, :, 0] = ihs_image[:, :, 0] / 180.0 * 255.0  # Hue
    rgb_image = cv2.cvtColor(ihs_image, cv2.COLOR_HSV2RGB)
    return rgb_image

def fuse_images(optical_image, radar_image):
    # Convertir l'image optique en IHS
    ihs_image = rgb_to_ihs(optical_image)

    # Remplacer l'intensité par l'image radar
    ihs_image[:, :, 2] = radar_image

    # Convertir de nouveau en RGB
    fused_image = ihs_to_rgb(ihs_image)
    return fused_image

# Charger les images
optical_image = cv2.imread('/home/francis/Bureau/formation_Memoire/newOptical-mountain/150.png')
radar_image = cv2.imread('/home/francis/Bureau/formation_Memoire/newSAR-mountain/150.png', cv2.IMREAD_GRAYSCALE)

# S'assurer que les dimensions des images correspondent
if optical_image.shape[:2] != radar_image.shape[:2]:
    radar_image = cv2.resize(radar_image, (optical_image.shape[1], optical_image.shape[0]))

# Fusionner les images
fused_image = fuse_images(optical_image, radar_image)

# Afficher le résultat
cv2.imshow('Fused Image', fused_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
