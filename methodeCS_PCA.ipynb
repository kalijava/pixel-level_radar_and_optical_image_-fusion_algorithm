```python
import cv2
import numpy as np
from sklearn.decomposition import PCA

def load_and_preprocess_images(optical_path, radar_path):
    optical_image = cv2.imread(optical_path)
    radar_image = cv2.imread(radar_path, cv2.IMREAD_GRAYSCALE)

    # S'assurer que les dimensions des images correspondent
    if optical_image.shape[:2] != radar_image.shape[:2]:
        radar_image = cv2.resize(radar_image, (optical_image.shape[1], optical_image.shape[0]))

    return optical_image, radar_image

def apply_pca(optical_image, radar_image):
    # Redimensionner les images pour l'ACP
    optical_reshaped = optical_image.reshape(-1, 3)
    radar_reshaped = radar_image.flatten().reshape(-1, 1)

    # Combiner les deux images
    combined = np.hstack((optical_reshaped, radar_reshaped))

    # Appliquer l'ACP
    pca = PCA(n_components=3)
    pca_result = pca.fit_transform(combined)

    return pca_result

def reconstruct_image(pca_result, optical_shape):
    # Reconstituer l'image  partir des composantes principales
    reconstructed_image = pca_result.reshape(optical_shape[0], optical_shape[1], 3)
    return np.clip(reconstructed_image, 0, 255).astype(np.uint8)

def fuse_images(optical_path, radar_path):
    optical_image, radar_image = load_and_preprocess_images(optical_path, radar_path)
    
    # Appliquer l'ACP
    pca_result = apply_pca(optical_image, radar_image)

    # Reconstituer l'image fusionne
    fused_image = reconstruct_image(pca_result, optical_image.shape)

    return fused_image

# Chemins vers les images
optical_path = 'fusion3.png'
radar_path = 'fusion3.png'

# Fusionner les images
fused_image = fuse_images(optical_path, radar_path)

# Afficher le rsultat
cv2.imshow('Fused Image', fused_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
